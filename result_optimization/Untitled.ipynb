```python
import os
from PIL import Image
import numpy as np
```

```python
folder = "E:\\deeplearning\\Hepatocarcinomes\\SlidesClassification"
```

```python
images = [i for i in os.listdir(folder) if i.endswith("_PbMap8bit.png")]
```

```python
images
```

```python
img = np.asarray(Image.open(os.path.join(folder, images[1])))
print(img)
img_rgb = np.stack((img,)*3, axis=-1)
print(img_rgb)
```

```python
for i in range(img_rgb.shape[0]):
    for j in range(img_rgb.shape[1]):
        if img_rgb[i, j, 0] >= 150:
            img_rgb[i,j,0] = 255
            img_rgb[i,j,1] = 0
            img_rgb[i,j,2] = 0
print(img_rgb)
```

```python
img_new = Image.fromarray(img_rgb, 'RGB')
print(img_new)
img_new.save(os.path.join("E:\\deeplearning\\Hepatocarcinomes\\test", images[1].replace("points", "PbMp8bit")))
```

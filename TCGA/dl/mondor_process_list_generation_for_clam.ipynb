## 3 kinds of Mondor slides: 20x, 40x_level1, 40x_custom
### Using OpenSlide, luckly svs and ndpi have the same selected attributes
1. Check highest magnification: f.properties[openslide.PROPERTY_NAME_OBJECTIVE_POWER]
2. Check if level of downsample 2.0 exist: f.level_downsamples[f.get_best_level_for_downsample(2)]
3. Check if this level is the second one: f.get_best_level_for_downsample(2)

##  Chi squared test for all pathological features of 6 signatures
### Independency test of features

```python
# conda envrionment deepai
```

```python
import os
import pandas as pd
import numpy as np
# install dependency for pd to read xlsx
# !conda install -c anaconda xlrd
from scipy.stats import chi2_contingency
```

### Path and load the gene signature list

```python
path = "/media/visiopharm5/WDGold/deeplearning/MIL/CLAM/patch_reviewing"
signatures = []
for root, dirs, files in os.walk(path):
    for file in files:
        if file.endswith('.xlsx') and not file.startswith('_'):
            signatures.append(file[:-5])
print(signatures)
```

## Data pre-processing

```python
for i in range(len(signatures)):
    # load patch reviewing data
    df = pd.read_excel(os.path.join(path, signatures[i]+'.xlsx'))
    # keep only patches in tp slides
    df = df.loc[df['class']=='tp']
    print(df.shape)
    display(df.head(3))
    
    # sum up counts of all features within a signature
    sumup = df.iloc[:, 4:].sum()
    
    if i == 0:
        # list features
        name_col = list(sumup.index)
        # df to summarize the feature counts in all signatures
        df_res = pd.DataFrame(columns = ['signature'] + name_col)
#         display(df_res)
    else:
        sumup = sumup[df_res.columns[1:]] # reorder the index
        assert list(sumup.index) == name_col                 
    
    df_res.loc[len(df_res)] = [signatures[i]] + list(sumup.values)

df_res.set_index('signature',inplace=True)
display(df_res)
```

```python
df_res.to_csv(os.path.join(path, 'summary.csv'), index=True)
```

## Chi squared test

```python
# check variable independency by chi sqaured test, ValueError since there is 0
# chi2_contingency(observed=df_res)
```

```python
# remove feautures (columns) containing values less than 5
df_res[df_res>=5].dropna(axis=1)
```

```python
# Chi squared test after removing feautures containing values less than 5
chi2_contingency(df_res[df_res>=5].dropna(axis=1))
```

### Change the less than 5 condition on expected table instead of observed table

```python
# calculate expected table
expected =  np.outer(df_res.sum(axis=1),
                     df_res.sum()) / df_res.sum().sum()

expected = pd.DataFrame(expected)

expected.columns = df_res.columns
expected.index = df_res.index

display(expected)
```

```python
expected.to_csv(os.path.join(path, 'expected.csv'), index=True)
```

```python
# remove feautures (columns) containing values less than 5
df_res[expected>=5].dropna(axis=1)
```

```python
# Chi squared test after removing feautures containing values less than 5 on expected table 
chi2_contingency(df_res[expected>=5].dropna(axis=1))
```

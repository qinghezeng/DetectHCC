## Add the max, mean and std of test AUC to the summary.csv, only for tcga validation
#### For Mondor, we should use the best fold found in TCGA

```python
import os
import numpy as np
import pandas as pd
```

```python
root = "/media/visiopharm5/WDGold/deeplearning/MIL/CLAM/eval_results_349_tumor_masked"
target = "EVAL_tcga_hcc_tumor-masked_349_6G_Interferon_Gamma_cv_highvsrest_622_MIL_50_s1_cv"
path = os.path.join(root, target)
```

```python
df_res = pd.read_csv(os.path.join(path, "summary.csv"), index_col=0)
display(df_res)
df_res.loc[len(df_res)] = ['best', df_res['test_auc'].max(), np.nan]
df_res.loc[len(df_res)] = ['mean', df_res.loc[:9, 'test_auc'].mean(), np.nan]
df_res.loc[len(df_res)] = ['std', df_res.loc[:9, 'test_auc'].std(), np.nan]
print(df_res.shape)
display(df_res)
df_res.to_csv(os.path.join(path, "summary.csv"))
```

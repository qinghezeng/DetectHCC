```python
# Gene signatures in Sangro's paper
root = "/media/visiopharm5/WDGold/deeplearning/MIL/CLAM"
if core_sample:
    #tcga
#     path_label = "/media/visiopharm5/WDGold/deeplearning/Hepatocarcinomes/TCGA/clust_res_final"
    #mondor
    path_label = "/media/visiopharm5/WDGold/deeplearning/Hepatocarcinomes/TCGA/clust_res_final"
    flabel = "mondor_sample_clusters_"+ gene_signature +"_zscore_hc_ward.D2_euclidean_3_reorder.csv"
else:
    # tcga
#     path_label = "/media/visiopharm5/WDGold/deeplearning/Hepatocarcinomes/TCGA/validation"
    # mondor
    path_label = "/media/visiopharm5/WDGold/deeplearning/Hepatocarcinomes/TCGA/heatmap_nanostring_139"
    flabel = "sample_clusters_"+ gene_signature +"_zscore_hc_ward.D2_euclidean_3_reorder.csv"
    
# ffname = "results/process_list_edited (copy).csv" # 20x + 40x
# merge = "outer"

# gene_signature = "Inflammatory" # Inflammatory, Gajewski_13G_Inflammatory, 6G_Interferon_Gamma, 
#                                         # Interferon_Gamma_Biology, T-cell_Exhaustion, Ribas_10G_Interferon_Gamma

# flabel = "tcga_sample_clusters_"+ gene_signature +"_zscore_hc_ward.D2_euclidean_3_reorder.csv"
# output = "dataset_csv/tcga_hcc_354_" + gene_signature
# tcga
# flabel = "mondor_sample_clusters_"+ gene_signature +"_zscore_hc_ward.D2_euclidean_3_reorder.csv"
output = "dataset_csv/mondor_hcc_258_" + gene_signature


if highvsrest:
    output = output + "_highvsrest"
elif lowvsrest:
    output = output + "_lowvsrest"

if core_sample:
    output = output + "_core.csv"
else:
    output = output + ".csv"
```

```python
# Load clustering results
labels = pd.read_csv(os.path.join(path_label, flabel))
# Load CSV auto-generated by CLAM segmentation -- for full slide name
df = pd.read_csv("/media/visiopharm5/WDGold/deeplearning/Hepatocarcinomes/TCGA/data/reference_mrna2slides.csv", index_col=0)
```

```python
# Data vis
print(labels.shape)
display(labels.head(5))

print(df.shape)
display(df.head(5))
```

```python
for i in range(df.shape[0]):
    if df.loc[i, 'slide_id'].startswith("HMNT0505_bis") or df.loc[i, 'slide_id'].startswith("HMNT0578 "):
        df.drop(i, inplace=True)
print(df.shape)
display(df.head(5))
```

```python
# Re-format sample name: (only for tcga, remove "X" before all numeric names and) replace "." with "-"
for i in range(labels.shape[0]):
#     labels.iloc[i,0] = labels.iloc[i,0][-8:-1].replace(".", "-")
    labels.iloc[i,0] = labels.iloc[i,0].replace(".", "-")
print(labels.shape)
display(labels.head(5))
```

```python
if highvsrest:
    labels.replace("Cluster Median", "Cluster Median + Low", inplace=True)
    labels.replace("Cluster Low", "Cluster Median + Low", inplace=True)
elif lowvsrest:
    labels.replace("Cluster High", "Cluster High + Median", inplace=True)
    labels.replace("Cluster Median", "Cluster High + Median", inplace=True)
display(labels.head(5))
```

```python
if core_sample:
    for i in range(labels.shape[0]): # redefine "core" sample
        if labels.iloc[i, 1] == labels.iloc[i, 2]:
            labels.iat[i, 4] = True
    
    labels = labels.drop(labels[~labels.core_sample].index).drop(columns=["neighbor", "sil_width", "core_sample"])
    print(labels.shape)
    display(labels.head(5))
```

```python
print(labels.shape)
display(labels.head(5))
```

```python
# # Build a new df
# df1 = df.filter(items=["slide_id"], axis="columns")
# df1["case_id"] = ""
# df1["Sample"] = ""
# for i in range(df.shape[0]):
#     df1.iloc[i,0] = df.iloc[i,0][:-4] # remove ".svs" at the end
#     df1.iloc[i,1] = df.iloc[i,0][0:12] # case id
#     df1.iloc[i,2] = df.iloc[i,0][8:15] # sample id which used to match the label df
# print(df1.shape)
# display(df1.head(5))
```

```python
# Matching the two df on sample id
# results = pd.merge(labels, df1, on="Sample", how="inner") # keep the matched duplicates (any case using merge 
                                                          # function) and preserve the left order
results = pd.merge(labels, df, left_on="Sample", right_on="Nanostring_id", how="inner")
print(results.shape)
display(results.head(5))
```

```python
results.drop(columns="case_id", inplace=True)
results.rename(columns={"Nanostring_id": "case_id"}, inplace=True)
print(results.shape)
display(results.head(5))
```

```python
# Reconstruct the results
results = results.filter(items=["case_id", "slide_id", "Cluster"], axis="columns")
results.rename(columns={"Cluster":"cluster"}, inplace=True)
print(results.shape)
display(results.head(5))
```

```python
# Export
results.to_csv(os.path.join(root, output))
```

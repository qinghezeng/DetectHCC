```python
import os
import numpy as np
import pandas as pd
import csv
```

```python
root = "/media/visiopharm5/WDGold/deeplearning/MIL/CLAM/eval_results"
target = "EVAL_mondor_hcc_tumor_258_Ribas_10G_Interferon_Gamma_cv_highvsrest_00X_CLAM_50_s1_cv"
path = os.path.join(root, target)
```

### Find all the csvs named "fold_k_optimal_tcga.csv"

```python
k = 0
for roots, dirs, files in os.walk(path):
    for file in files:
        if file.startswith("fold_") and len(file)== 23:
            k = k + 1
print(k) # folds
```

### Load the csvs and summerize

```python
# cases
df_res = pd.DataFrame(columns=['fold', 'consistent predictions', # 'consistent predictions (ratio)', 
                               'consistent predictions in Cluster High', 
#                                'consistent predictions in Cluster High (ratio)',
                               'consistent predictions in Cluster High true positive', 
#                                'consistent predictions in Cluster High true positive (ratio)',
                               'single predictions', # 'single predictions (ratio)', 
                               'single predictions in Cluster High', 
#                                'single predictions in Cluster High (ratio)',
                               'single predictions in Cluster High true positive' 
#                                'single predictions in Cluster High true positive (ratio)'
                              ])

for fold in range(k):
    df = pd.read_csv(os.path.join(path, "fold_"+str(fold)+"_optimal_tcga.csv"))
    print(df.shape)
    display(df.head(5))
    
    consistent_predictions = df['consistent_prediction'].sum()/2 # sum() doesn't work with a column including NaN
#     consistent_predictions_ratio = consistent_predictions/df.shape[0]
    consistent_predictions_in_Cluster_High = (df['consistent_prediction'] & (df.Y==1.0)).sum()/2
#     consistent_predictions_in_Cluster_High_ratio = consistent_predictions_in_Cluster_High/sum((df.Y==1.0))
    consistent_predictions_in_Cluster_High_true_positive = (df['consistent_prediction'] & (df.Y==1.0) & df['true_prediction']).sum()/2
#     consistent_predictions_in_Cluster_High_true_positive_ratio = consistent_predictions_in_Cluster_High_true_positive/sum((df.Y==1.0) and true_prediction)/2)
    single_slides = sum(df['consistent_prediction'].isna())
    single_predictions = consistent_predictions + single_slides
    single_predictions_in_Cluster_High = consistent_predictions_in_Cluster_High + (df['consistent_prediction'].isna() & (df.Y==1.0)).sum()
    single_predictions_in_Cluster_High_true_positive = consistent_predictions_in_Cluster_High_true_positive + (df['consistent_prediction'].isna() & (df.Y==1.0) & df['true_prediction']).sum()
    
    
    
    df_res = df_res.append({'fold': fold, 'consistent predictions': consistent_predictions, 
                            'consistent predictions in Cluster High': consistent_predictions_in_Cluster_High,
                            'consistent predictions in Cluster High true positive': consistent_predictions_in_Cluster_High_true_positive,
                            'single predictions': single_predictions,
                            'single predictions in Cluster High': single_predictions_in_Cluster_High, 
                            'single predictions in Cluster High true positive': single_predictions_in_Cluster_High_true_positive}, ignore_index=True)
```

### Visualize the results and export

```python
print(df_res.shape)
display(df_res)
df_res.to_csv(os.path.join(path, "consistent_summary_optimal_tcga.csv"), index=False)
```

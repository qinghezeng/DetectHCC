```python
import os
import pandas as pd
```

```python
path = '/media/visiopharm5/WDGold/deeplearning/Hepatocarcinomes/TCGA/data/clinicopathologic/tcga_raw/cbioportal_firehose'
fref = '/media/visiopharm5/WDGold/deeplearning/MIL/CLAM/dataset_csv/tcga_hcc_349_Inflammatory_highvsrest.csv'
```

```python
# load ref including all the tcga used cases
df_ref = pd.read_csv(fref, index_col=0)
print(df_ref.shape) # (349, 3)
display(df_ref.head(3))
```

```python
# extract tcga case list
list_tumor = df_ref['case_id'].values
print(len(set(list_tumor))) # 336 unique cases
print(list_tumor[:4])
```

```python
# load tcga data
df = pd.read_csv(os.path.join(path, 'data_bcr_clinical_data_patient.txt'), '\t')
print(df.shape) # (381, 74)
display(df.head(5))
```

```python
# df_res as results and df as source data
df_res = df.copy(deep=True)
print(df_res.shape) # (381, 74)
display(df_res.head(3))
```

```python
# filter df by used tcga case list

# make sure no used cases missing in df. Real data is from the row 4
assert len(set(list_tumor) - set(list(df_res['Patient Identifier'].values)[4:]))==0

df_res = df_res[df_res['Patient Identifier'].isin(list_tumor)].reset_index(drop=True)
print(df_res.shape) # (336, 74)
display(df_res.head(3))
```

## Variable analysis

```python
# available data. In the column 'Surgical Margin Resection Status', RX should be treated as not availale
list_variables = ['Sex', 'Diagnosis Age', 'History hepato carcinoma risk factor', 'AFP At Procurement',
                 'Neoplasm Histologic Grade', 'Vascular Invasion', 'Surgical Margin Resection Status',
                 'Liver fibrosis ishak score category']
series_available = (len(df_res) - ((df_res=='[Not Available]') | (df_res=='RX')).sum())[list_variables]
display(series_available)
```

```python
# available data (percentage corrected to 2 decimal places)
round(series_available / len(df_res) * 100, 2)
```

```python
# dtypes
df_res[list_variables].dtypes
```

```python
# Sex
# print(df.loc[:3,'Sex'])
# print(set(df.loc[4:,'Sex'].values))
print('Sex: {}'.format(set(df_res['Sex'].values)))
n_male = len(df_res[df_res['Sex']=='Male'])
n_female = len(df_res[df_res['Sex']=='Female'])
print('Male: {} ({}%)'.format(n_male, round(n_male / len(df_res) * 100, 2)))
print('Female: {} ({}%)'.format(n_female, round(n_female / len(df_res) * 100, 2)))
```

```python
# Age
# print(df.loc[:3,'Diagnosis Age'])
# print(set(df.loc[4:,'Diagnosis Age'].values))
print('Age')
print('Median: {}'.format(df_res[df_res['Diagnosis Age']!='[Not Available]']['Diagnosis Age'].median()))
print('Min: {}'.format(df_res[df_res['Diagnosis Age']!='[Not Available]']['Diagnosis Age'].min()))
print('Max: {}'.format(df_res[df_res['Diagnosis Age']!='[Not Available]']['Diagnosis Age'].max()))
```

```python
# History hepato carcinoma risk factor
# print(df.loc[:3,'History hepato carcinoma risk factor'])
# print(set(df.loc[4:,'History hepato carcinoma risk factor'].values))
print('History hepato carcinoma risk factor: {}'.format(set(df_res['History hepato carcinoma risk factor'])))
# Hepatitis B
n_hbv = df_res['History hepato carcinoma risk factor'].str.contains('Hepatitis B').sum()
print('Hepatitis B: {} ({}%)'.format(n_hbv, round(n_hbv/len(df_res)*100, 2)))
# Hepatitis C
n_hcv = df_res['History hepato carcinoma risk factor'].str.contains('Hepatitis C').sum()
print('Hepatitis C: {} ({}%)'.format(n_hcv, round(n_hcv/len(df_res)*100, 2)))
# Alcohol consumption
n_alcohol = df_res['History hepato carcinoma risk factor'].str.contains('Alcohol consumption').sum()
print('Alcohol consumption: {} ({}%)'.format(n_alcohol, round(n_alcohol/len(df_res)*100, 2)))
# Non-Alcoholic Fatty Liver Disease
n_nash = df_res['History hepato carcinoma risk factor'].str.contains('Non-Alcoholic Fatty Liver Disease').sum()
print('Non-Alcoholic Fatty Liver Disease: {} ({}%)'.format(n_nash, round(n_nash/len(df_res)*100, 2)))
# No History of Primary Risk Factors
n_nofactor = df_res['History hepato carcinoma risk factor'].str.contains('No History of Primary Risk Factors').sum()
print('No History of Primary Risk Factors: {} ({}%)'.format(n_nofactor, round(n_nofactor/len(df_res)*100, 2)))
```

```python
# AFP At Procurement
# print(df.loc[:3,'AFP At Procurement'])
# print(set(df.loc[4:,'AFP At Procurement'].values))
print('AFP At Procurement')
# check if all int: yes
assert len(set([len(i) for i in df_res[df_res['AFP At Procurement']!='[Not Available]']['AFP At Procurement'].str.split('.')]))==1
print('Median: {}'.format(df_res[df_res['AFP At Procurement']!='[Not Available]']['AFP At Procurement'].median()))
print('Min: {}'.format(df_res[df_res['AFP At Procurement']!='[Not Available]']['AFP At Procurement'].min()))
print('Max: {}'.format(df_res[df_res['AFP At Procurement']!='[Not Available]']['AFP At Procurement'].max()))
```

```python
# Neoplasm Histologic Grade
# print(df.loc[:3,'Neoplasm Histologic Grade'])
# print(set(df.loc[4:,'Neoplasm Histologic Grade'].values))
print('Neoplasm Histologic Grade: {}'.format(set(df_res['Neoplasm Histologic Grade'])))
n_g1 = len(df_res[df_res['Neoplasm Histologic Grade']=='G1'])
print('G1: {} ({}%)'.format(n_g1, round(n_g1/len(df_res)*100, 2)))
n_g2 = len(df_res[df_res['Neoplasm Histologic Grade']=='G2'])
print('G2: {} ({}%)'.format(n_g2, round(n_g2/len(df_res)*100, 2)))
n_g3 = len(df_res[df_res['Neoplasm Histologic Grade']=='G3'])
print('G3: {} ({}%)'.format(n_g3, round(n_g3/len(df_res)*100, 2)))
n_g4 = len(df_res[df_res['Neoplasm Histologic Grade']=='G4'])
print('G4: {} ({}%)'.format(n_g4, round(n_g4/len(df_res)*100, 2)))
```

```python
# Vascular Invasion
# print(df.loc[:3,'Vascular Invasion'])
# print(set(df.loc[4:,'Vascular Invasion'].values))
print('Vascular Invasion: {}'.format(set(df_res['Vascular Invasion'])))
n_micro = len(df_res[df_res['Vascular Invasion']=='Micro'])
print('Micro: {} ({}%)'.format(n_micro, round(n_micro/len(df_res)*100, 2)))
n_macro = len(df_res[df_res['Vascular Invasion']=='Macro'])
print('Macro: {} ({}%)'.format(n_macro, round(n_macro/len(df_res)*100, 2)))
```

```python
# Surgical Margin Resection Status
# print(df.loc[:3,'Surgical Margin Resection Status'])
# print(set(df.loc[4:,'Surgical Margin Resection Status'].values))
print('Surgical Margin Resection Status: {}'.format(set(df_res['Surgical Margin Resection Status'])))
n_r0 = len(df_res[df_res['Surgical Margin Resection Status']=='R0'])
print('R0: {} ({}%)'.format(n_r0, round(n_r0/len(df_res)*100, 2)))
n_r1r2 = len(df_res[(df_res['Surgical Margin Resection Status']=='R1') | 
                    (df_res['Surgical Margin Resection Status']=='R2')])
print('R1R2: {} ({}%)'.format(n_r1r2, round(n_r1r2/len(df_res)*100, 2)))
```

```python
# Liver fibrosis ishak score category
# print(df.loc[:3,'Liver fibrosis ishak score category'])
# print(set(df.loc[4:,'Liver fibrosis ishak score category'].values))
print('Liver fibrosis ishak score category: {}'.format(set(df_res['Liver fibrosis ishak score category'])))
n_fibrosis_012 = len(df_res[(df_res['Liver fibrosis ishak score category']=='0 - No Fibrosis') | 
                            (df_res['Liver fibrosis ishak score category']=='1,2 - Portal Fibrosis')]) 
print('012: {} ({}%)'.format(n_fibrosis_012, round(n_fibrosis_012/len(df_res)*100, 2)))
n_fibrosis_34 = len(df_res[(df_res['Liver fibrosis ishak score category']=='3,4 - Fibrous Speta')])
print('34: {} ({}%)'.format(n_fibrosis_34, round(n_fibrosis_34/len(df_res)*100, 2)))
n_fibrosis_56 = len(df_res[(df_res['Liver fibrosis ishak score category']=='5 - Nodular Formation and Incomplete Cirrhosis') | 
                           (df_res['Liver fibrosis ishak score category']=='6 - Established Cirrhosis')])
print('56: {} ({}%)'.format(n_fibrosis_56, round(n_fibrosis_56/len(df_res)*100, 2)))
```

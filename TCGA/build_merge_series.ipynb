## Merge TCGA and Mondor series for intergation (clustered heatmap)
###  Input: TCGA and Mondor data with full genes after Z-Score
### Output: merged dataset

```python
import os
import pandas as pd
```

```python
path = "/media/visiopharm5/WDGold/deeplearning/Hepatocarcinomes/TCGA/merge_series"
path_tcga = "/media/visiopharm5/WDGold/deeplearning/Hepatocarcinomes/TCGA/processed"
path_mondor = "/media/visiopharm5/WDGold/deeplearning/Hepatocarcinomes/TCGA/heatmap_nanostring_155"

file_tcga = "fpkm_final_log2-zscore_CCL3L1-noDEFB134.csv"
file_mondor = "mondor_final_log2-zscore-centering.csv"
```

### Load tcga and mondor dataset
#### tcga gene list: clean full genes (733)
#### mondor gene list: real full genes (747)

```python
df_tcga = pd.read_csv(os.path.join(path, file_tcga), sep="\t", index_col=0)
print(df_tcga.shape)
display(df_tcga.head(5))
```

```python
df_mondor = pd.read_csv(os.path.join(path, file_mondor), sep="\t", index_col=0)
print(df_mondor.shape)
display(df_mondor.head(5))
```

### Merge the two dataset only keeping overlapping genes

```python
df_res = pd.concat([df_mondor, df_tcga], join="inner", axis=1)
print(df_res.shape)
display(df_res.head(5))
```

### Export

```python
df_res.to_csv(os.path.join(path, "final_log2-zscore-centering_merged_mondor_tcga.csv"), sep='\t', index=True)
```

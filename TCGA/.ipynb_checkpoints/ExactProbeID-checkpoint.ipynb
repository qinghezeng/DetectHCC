## Extra ProbeID list to request information from web E!Emsembl
### Output: A csv file

Requested info:

1. **Dataset**: Human genes (GRCh38.p13)
    
2. **Filters (Input)**: splitted ProbeID (***the csv file generated by this script***, *RefSeq mRNA ID*)
    
3. **Attributes (Output)**: 
    
    1. *Gene stable ID version* (NIH GDC data, non-unique)
    
    2. *Gene name* (unique official name, but cBioPortal and Firehose use Alias and former name)
            
    3. *NCBI gene ID* (unique, ID used by cBioPortal and Firehose)
            
    4. *RefSeq mRNA ID* (ID used by Mondor)
            
    (Enable *Unique results only*, click *GO* to download file directly)

```python
import pandas as pd
```

### Load and visualize data 

```python
data = pd.read_csv('E:\\deeplearning\\Hepatocarcinomes\\TCGA\\ProbeAnnotations_NS_IO_360_v1.0_clean.csv')
```

```python
print(data.shape)
display(data.head(8))
```

### Exact proper ProbeID

```python
ID = pd.DataFrame([id.split(".")[0] for id in data["ProbeID"]])
ID.columns = ["ProbeID"]
```

```python
print(ID.shape)
display(ID.head(8))
```

### Export as csv

```python
ID.to_csv("E:\\deeplearning\\Hepatocarcinomes\\TCGA\\ProbeID.csv", sep='\t', index=False, header= False)
```

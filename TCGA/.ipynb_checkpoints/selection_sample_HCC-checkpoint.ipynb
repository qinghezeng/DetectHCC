```python
# Filter the HCC slides
df_slides = pd.DataFrame(slides,columns=["names"])
print(df_slides.shape) # (379, 1)
wanted=[]
for slideid in (set_case_firebrowse.intersection(set_case_gt.intersection(set_case_slides))):
    wanted.extend(df_slides[df_slides.names.str.startswith(slideid)].index)
# unwanted
df_slides = df_slides.filter(items=wanted, axis="index")
df_slides.reset_index(drop=True, inplace=True)
```

```python
unwanted=[]
for slideid in (set_case_firebrowse.intersection(set_case_gt.intersection(set_case_slides))):
    unwanted.extend(df_slides.names[df_slides.names.str.startswith(slideid)])
df_slides_case = pd.DataFrame(i.split(".")[0][:-1] for i in unwanted)
df_slides['duplicated']=df_slides_case.duplicated(keep=False)
df_slides.replace(False, "", inplace=True)
df_slides
# Export list which in TCGA-LIHC project but not in HCC list from Julien
df_slides.to_csv("E:\\deeplearning\\Hepatocarcinomes\\TCGA\\slidesHCC.csv", index=False, sep='\t')
```

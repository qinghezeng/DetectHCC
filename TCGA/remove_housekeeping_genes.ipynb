```python
import pandas as pd
```

```python
data =  pd.read_csv("E:\\deeplearning\\Hepatocarcinomes\\TCGA\\processed\\firehose_filtered.csv", sep="\t")
print(data.shape)
display(data.head(5))
```

```python
mondor =  pd.read_csv("E:\\deeplearning\\Hepatocarcinomes\\TCGA\\ProbeAnnotations_NS_IO_360_v1.0_clean.csv", sep="\t")
print(mondor.shape)
display(mondor.head(5))
```

```python
id_hk = mondor[mondor["Control.Type"]=="Housekeeping"].ProbeID.values.tolist()
print(len(id_hk))
print(id_hk)
```

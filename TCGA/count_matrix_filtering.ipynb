### Load count matrix and subset the fpkm gene list.
### Check missing data.
### Saved as CSV.

```python
import pandas as pd
import numpy as np
```

```python
root = "E:\\deeplearning\\Hepatocarcinomes\\TCGA\\processed\\"
```

```python
count_matrix = pd.read_csv(root + 'count_matrix_TCGA_424s.txt', sep='\t', index_col=0)
fpkm = pd.read_csv(root + 'fpkm_final_log2-median-centering.csv', sep='\t', index_col=0)
```

```python
print("count_matrix:")
display(count_matrix.head(5))
print(count_matrix.shape) # (18441, 424)
print("")
print("Used full fpkm:")
display(fpkm.head(5))
print(fpkm.shape) # (733, 340)
```

```python
count_matrix = count_matrix.filter(fpkm.index.values, axis = "index")
display(count_matrix.head(5))
print(count_matrix.shape) # (733, 424)
```

```python
count_matrix = count_matrix.filter(fpkm.columns.values)
display(count_matrix.head(5))
print(count_matrix.shape) # (733, 340)
```

```python
count_matrix.to_csv(root + 'count_matrix_filtered_full.csv', index=True, sep='\t')
```

```python
# Missing elements
print("Number of NaN or missing data: " + str(count_matrix.isnull().sum().sum())) # 0, should not be any missing data
print("Number of 0: " + str((count_matrix==0).sum().sum())) # 10088, 0 doesn't mean missing
```

```python
# Missing names of row or col
print("Missing values in gene name: " + str(count_matrix.index.isna().sum())) # 0 
print("Missing values in sample id: " + str(count_matrix.columns.isna().sum())) # 0
```

```python
# How many all-NA rows and their indices
print("All-zero rows: " + str(((count_matrix.T==0).sum()==340).sum())) # 1, normal. Not expressed at all.
display(count_matrix.iloc[np.where((count_matrix.T==0).sum()==340)]) # also all-zero in fpkm_TCGA
print("All-zero cols: " + str(((count_matrix==0).sum()==733).sum())) # 0
```

```python
display(count_matrix.T.describe())
print("Minimum of the data: "+str(count_matrix.min().min())) # 0
```

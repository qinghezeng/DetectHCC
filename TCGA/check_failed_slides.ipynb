```python
import os
import csv
import pandas as pd
import shutil
```

```python
i=0
failed=[]
failed_dirs=[]
failed_size=[]
for root, dirs, files in os.walk("E:\\deeplearning\\Hepatocarcinomes\\TCGA\\diagnostic_slides", topdown=False):
    for file in files:
        if file.endswith(".partial"):
            failed.append(file) #379
            failed_dirs.append(root)
            failed_size.append(os.path.getsize(os.path.join(root,file)))
            i=i+1
print(i)
print(min(failed_size))
```

```python
for dir in failed_dirs:
    shutil.rmtree(dir)
    shutil.move(dir.replace("E:\\deeplearning\\Hepatocarcinomes\\TCGA\\", "Y:\\"), dir)
```

```python
i=0
succeeded=[]
succeeded_size=[]
for root, dirs, files in os.walk("E:\\deeplearning\\Hepatocarcinomes\\TCGA\\diagnostic_slides", topdown=False):
    for file in files:
        if file.endswith(".svs"):
            succeeded.append(file) #212
            succeeded_size.append(os.path.getsize(os.path.join(root,file)))
            i=i+1
print(i)
print(max(succeeded_size))
```

```python
with open("E:\\deeplearning\\Hepatocarcinomes\\TCGA\\failed_diagn_slides\\failed_slide_name.csv", 'w', newline="") as myfile:
    wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
    for slide in failed:
        wr.writerow([slide])
```

```python
manifest = pd.read_csv('E:\\deeplearning\\Hepatocarcinomes\\TCGA\\diagnostic_slides\\gdc_manifest.2020-03-11.txt', sep="\t")
print(manifest.shape)
display(manifest.head(5))
```

```python
for slide in succeeded:
    manifest.drop(manifest[manifest.filename==slide].index, inplace=True)
print(manifest.shape)
```

```python
manifest.to_csv('E:\\deeplearning\\Hepatocarcinomes\\TCGA\\diagnostic_slides\\gdc_manifest_failed_212.txt', sep="\t", index=False)
```

```python
i=0
failed1=[]
for root, dirs, files in os.walk("Y:\\diagnostic_slides", topdown=False):
    for file in files:
        if file.endswith("svs"):
            failed1.append(file) #212
            i=i+1
print(i)
print(list(set(failed1)-set(succeeded)))
```

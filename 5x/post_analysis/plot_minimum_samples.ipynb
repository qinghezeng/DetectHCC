```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv('E:\\deeplearning\\Hepatocarcinomes\\models\\5x\\test_dataAmount\\minimumsample.csv', delimiter='\t', decimal=',')
df2 = pd.read_csv('E:\\deeplearning\\Hepatocarcinomes\\models\\5x\\test_dataAmount\\minimumsamplemean.csv', delimiter=';', decimal=',')
```

```python
plt.figure(figsize=(16, 6))
plt.ylim(0,1)
sns.set_style("ticks")
sns.set_context("paper")
g = sns.boxplot(x='Amount', y='AUC', data=df1, notch=False, showmeans=True, meanline=True, palette='Set3')
g.set_xlabel("X Label",fontsize=15)
g.set_ylabel("Y Label",fontsize=15)
g.tick_params(labelsize=11)
g.set_xticklabels(g.get_xticklabels(),rotation=30)
sns.swarmplot(x='Amount', y='AUC', data=df1,
              size=2, color=".3", linewidth=0)
g = sns.swarmplot(x='Amount', y='AUC', data=df1, color=".25", size=6)
# g.set_xscale("log")
# plt.xlim(1,500000)
g=plt.plot(range(len(g.get_xticklabels())), df2['AUC'].values, 'r') #the x axe for boxplot is int beginning from 0

# plt.savefig('D:\\qinghe\\figures\\final\\res2a_minimum_samples.jpg')
```

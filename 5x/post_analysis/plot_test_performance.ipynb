```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('E:\\deeplearning\\Hepatocarcinomes\\SlidesClassification\\vgg16_performance_test.csv', delimiter=';', decimal='.')
```

```python
# sns.lvplot(x="day", y="total_bill", data=df, palette="mako",
#            #hue = 'smoker',
#            width = 0.8,           # 
#            linewidth = 12,
#            scale = 'area',        #   linearexonentialarea
#            k_depth = 'proportion',  #   proportiontukeytrustworthy
#           )
# LV

plt.figure(figsize=(20, 15))
plt.ylim(0,1)
sns.set_style("ticks")
sns.set_context("paper")
g=sns.boxplot(order=["Sensitivity (TPR)","Specificity (TNR)","Precision (PPV)","Fall-Out (FPR)","Area under ROC (AUC)",
                   "Accuracy (ACC)","F1-Score","Intersection over Union (IoU)"], data=df, notch=True, meanline=True, 
                  width=0.7, showfliers=False, palette='Set3')
g.set_xlabel("Metrics",fontsize=20)
g.set_ylabel("Distribtuion",fontsize=20)
g.tick_params(labelsize=14)
g.set_xticklabels(g.get_xticklabels(),rotation=30)
sns.swarmplot(order=["Sensitivity (TPR)","Specificity (TNR)","Precision (PPV)","Fall-Out (FPR)","Area under ROC (AUC)",
                     "Accuracy (ACC)","F1-Score","Intersection over Union (IoU)"], data=df,color ='k',size = 3,alpha = 0.8)
plt.tight_layout()
plt.savefig('D:\\qinghe\\figures\\final\\res3a_test_metrics.jpg')
```

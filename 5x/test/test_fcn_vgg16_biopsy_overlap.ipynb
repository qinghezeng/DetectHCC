```python
def testGenerator(img,target_size = (64,64),flag_multi_class = True):
    for i in range(img.shape[0]):
        image = cv2.resize(img[i],target_size) #resize function only support float image
#         print(img.shape)
        image = np.reshape(image,image.shape+(1,)) if (not flag_multi_class) else image
        image = np.reshape(image,(1,)+image.shape)
#         print(img.shape)
        yield image
```

```python
for i in range(len(limg)):
    image = cv2.cvtColor(cv2.imread(limg[i], -1), cv2.COLOR_BGR2RGB)/255
    reflect = cv2.copyMakeBorder(image,64,64,64,64,cv2.BORDER_REFLECT)
    img = np.ndarray((64, 192, 192, 3))
    for j in range(8):
        for k in range(8):
            img[j*8+k,:,:,:] = reflect[k*64:k*64+192,j*64:j*64+192,:]
    del(image)
    del(reflect)
    
    preds = model.predict_generator(testGenerator(img),img.shape[0],verbose=1)
    
    mask = np.ndarray((512, 512))
    for i,item in enumerate(preds):
        res = item[:,1]
        print(res.shape)
        res = res.reshape(64, 64)*255
        res = cv2.resize(res, (192, 192))
        mask[i%64%8*64:(i%64%8+1)*64,i%64//8*64:(i%64//8+1)*64] = res[64:128,64:128]
        if (i%64 == 63):
            n = str(len(os.listdir(path2+'\\label_overlap_fcn')))
            cv2.imwrite(path2+'\\label_overlap_fcn\\'+n.zfill(5)+'.tiff',mask.astype(np.uint8))
```

```python
for i in range(len(limg)):
    image = cv2.cvtColor(cv2.imread(limg[i], -1), cv2.COLOR_BGR2RGB)/255
    img = np.ndarray((64, 64, 64, 3))
    for j in range(8):
        for k in range(8):
            img[j*8+k,:,:,:] = image[k*64:k*64+64,j*64:j*64+64,:]
    del(image)
    
    preds = model.predict_generator(testGenerator(img),img.shape[0],verbose=1)
    
    mask = np.ndarray((512, 512))
    for i,item in enumerate(preds):
        res = item[:,1]
        print(res.shape)
        res = res.reshape(64, 64)*255
        mask[i%64%8*64:(i%64%8+1)*64,i%64//8*64:(i%64//8+1)*64] = res[:,:]
        if (i%64 == 63):
            n = str(len(os.listdir(path2+'\\label_fcn')))
            cv2.imwrite(path2+'\\label_fcn\\'+n.zfill(5)+'.tiff',mask.astype(np.uint8))
```

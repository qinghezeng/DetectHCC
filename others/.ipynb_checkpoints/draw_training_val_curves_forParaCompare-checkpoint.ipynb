```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
loss_adam_lr3_64 = []
val_loss_adam_lr3_64 = []
acc_adam_lr3_64 = []
val_acc_adam_lr3_64 = []

with open('E:\\deeplearning\\Hepatocarcinomes\\models\\5x\\test_dataAmount\\50k\\adam_64_lr3_training.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            line_count += 1
        else:
            loss_adam_lr3_64.append(row['loss'])
            val_loss_adam_lr3_64.append(row['val_loss'])
            acc_adam_lr3_64.append(row['acc'])
            val_acc_adam_lr3_64.append(row['val_acc'])
            line_count += 1
            
loss_adam_lr3_64=np.array(loss_adam_lr3_64, dtype=np.float32)#to use matplotlib, has to change list to np.array, otherwise the y axis will be autoscaled
val_loss_adam_lr3_64=np.array(val_loss_adam_lr3_64, dtype=np.float32)
acc_adam_lr3_64=np.array(acc_adam_lr3_64, dtype=np.float32)
val_acc_adam_lr3_64=np.array(val_acc_adam_lr3_64, dtype=np.float32)
```

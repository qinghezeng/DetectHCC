### Functions

```python
# ts.tileIterator will start to extract patch from (0, 0).
# The part of patches overlapped with the ROI will be return, without padding.
# So the returned patches could be retangle of any size.
# ts.getSingleTile will return the first patch, without padding.
# So here we have to use ts.getRegion, to extract a square at a specific location.
def read_image(ts, left, top):
    ratio=ts.getMetadata()['magnification']/10
    im_roi, _ = ts.getRegion(
        region=dict(left=left, top=top, width=128*ratio, height=128*ratio, units='base_pixels'),
        scale=dict(magnification=10),
        format=large_image.tilesource.TILE_FORMAT_NUMPY
    )
    im_roi = im_roi[:,:,:3]
    plt.imshow(im_roi)
    print(im_roi.shape)

    return im_roi
```

```python
# Generate mask
def generate_mask(im_src):
    labeled, _ = tissue_detection.get_tissue_mask(
        im_src, deconvolve_first=False,
        n_thresholding_steps=1, sigma=0., min_size=3)
    
    mask_out = resize(
        labeled == 0, output_shape=np.array(im_src).shape[:2],
        order=0, preserve_range=True) == 1

    f, ax = plt.subplots(1, 3, figsize=(10, 10))
    ax[0].imshow(im_src)
    ax[1].imshow(labeled) # all tissue regions
    ax[2].imshow(mask_out) # for normalization
    plt.show()

    return(mask_out)
```

```python
# Color normalization
def color_nor(method, im_target, im_src, src_mask=None):
    
    if method == "reinhard":
        target_mean, target_std = histomicstk.preprocessing.color_conversion.lab_mean_std(im_target, mask_out=None)
        if src_mask is None:
            im_normalized = histomicstk.preprocessing.color_normalization.reinhard(
                im_src, target_mu=target_mean, target_sigma=target_std)
            f, ax = plt.subplots(1, 3, figsize=(10, 10))
            ax[0].imshow(im_target)
            ax[1].imshow(im_src)
            ax[2].imshow(im_normalized)
            plt.show()
        else:
            im_normalized = histomicstk.preprocessing.color_normalization.reinhard(
                im_src, target_mu=target_mean, target_sigma=target_std, mask_out=src_mask)

            f, ax = plt.subplots(1, 4, figsize=(10, 10))
            ax[0].imshow(im_target)
            ax[1].imshow(im_src)
            ax[2].imshow(im_normalized)
            ax[3].imshow(src_mask)
            plt.show()
            
    elif method == "macenko":
        stain_unmixing_routine_params = {
            'stains': ['hematoxylin', 'eosin'],
            'stain_unmixing_method': 'macenko_pca',
            }
        if src_mask is None:
            im_normalized = deconvolution_based_normalization(
                im_src, im_target=im_target,
                stain_unmixing_routine_params=stain_unmixing_routine_params)
            f, ax = plt.subplots(1, 3, figsize=(10, 10))
            ax[0].imshow(im_target)
            ax[1].imshow(im_src)
            ax[2].imshow(im_normalized)
            plt.show()
        else:
            im_normalized = deconvolution_based_normalization(
                im_src, im_target=im_target,
                stain_unmixing_routine_params=stain_unmixing_routine_params,
                mask_out=src_mask)

            f, ax = plt.subplots(1, 4, figsize=(10, 10))
            ax[0].imshow(im_target)
            ax[1].imshow(im_src)
            ax[2].imshow(im_normalized)
            ax[3].imshow(src_mask)
            plt.show()
        
    return (im_normalized)
```
